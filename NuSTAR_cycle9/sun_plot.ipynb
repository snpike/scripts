```python
import numpy as np
import ligo.skymap.plot
from matplotlib import pyplot as plt
from astropy import units as u
from astropy.coordinates import get_sun
from astropy.coordinates import angular_separation
from astropy.coordinates import SkyCoord
from astropy.time import Time
from matplotlib.patches import Circle
```

```python
ra = []
dec = []
unknown = []
datetime = []
with open('/Users/sean/scripts/NuSTAR_cycle9/trigger_cat.txt') as file:
    for line in file:
        ra.append(float(line.split()[4]))
        dec.append(float(line.split()[5]))
        unknown.append(line.split()[3]=='Unknown')
        temp_date = (line.split()[1]).split('/')
        datetime.append('20' + temp_date[0] +'-' + temp_date[1] + '-' + temp_date[2] + ' ' + line.split()[2])

c_gcrs = SkyCoord(ra = np.array(ra)*u.degree, dec= np.array(dec)*u.degree, frame='icrs').gcrs
sun_pos = get_sun(Time(val = datetime))
sun_separation = angular_separation(c_gcrs.ra, c_gcrs.dec, sun_pos.ra, sun_pos.dec)
constrained = (sun_separation * 180/np.pi).value < 40
np.sum(constrained)/len(constrained)
```

```python
# suntime = Time(val = np.arange(-695, 3781, 1) + 58849, format='mjd')
suntime = Time(val = np.arange(80, 80+365, 1) + 58849, format='mjd')
sunpath = get_sun(suntime)
suntime
```

```python
plt.figure(figsize=(16,9))
ax = plt.axes(projection='astro degrees aitoff')
ax.grid()
ax.scatter(c_gcrs.ra.degree[~constrained], c_gcrs.dec.degree[~constrained], transform=ax.get_transform('world'), color = 'C0', s=75.0)
ax.scatter(c_gcrs.ra.degree[constrained], c_gcrs.dec.degree[constrained], transform=ax.get_transform('world'), color = 'red', s=200.0, marker='x')
ax.plot(sunpath.ra.degree, sunpath.dec.degree, transform=ax.get_transform('world'), color = 'C1', lw = 5.0)

# ax_zoom_circle = plt.axes(
#     [0.55, 0.1, 0.6, 0.6],
#     projection='galactic degrees zoom',
#     center='0d +0d',
#     radius='40 deg')

galactic_center_gcrs = SkyCoord(l =  0*u.degree, b= 0*u.degree, frame='galactic').gcrs
sun_to_sag = (angular_separation(sunpath.ra, sunpath.dec, galactic_center_gcrs.ra, galactic_center_gcrs.dec) * 180/np.pi).value
print(Time(val = np.arange(80, 80+365, 1) + 58849, format='mjd')[np.argmin(sun_to_sag)])


ax.add_patch(Circle((sunpath.ra.degree[np.argmin(sun_to_sag)], sunpath.dec.degree[np.argmin(sun_to_sag)]), 40,  transform=ax.get_transform('world'), color='C1', alpha=0.4))
plt.tight_layout()
plt.savefig('/Users/sean/scripts/NuSTAR_cycle9/skymap.pdf')
```
