```python
import matplotlib.pyplot as plt
import numpy as np

h = 4.136e-18
c = 3e8

def planck(E, kT, beta = 0.0):
    beta_term = np.sqrt((1+beta)/(1-beta))
    return 2*np.power(E,3)*np.power(beta_term, 4)/(np.power(h, 3)*np.square(c)*(np.exp(E*beta_term/(kT)) - 1.0))
```

```python
energy = np.arange(3, 78, 0.04)
plt.plot(energy, planck(energy, 3.0), label = 'Classical')
plt.plot(energy, planck(energy, 3.0, beta=-0.25), label = 'Relativistic')
plt.legend()
plt.loglog()
plt.show
```
