```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib as mpl
from scipy.optimize import curve_fit
from scipy import odr
import seaborn as sns
import scipy.signal as sig
from astropy.table import Table
from astropy.io import ascii

sns.set_context('poster')
sns.set_style('ticks')
sns.set_palette('colorblind')

#distance (3Mpc) in cm from Koribalski+ 2004, H=70km/s/Mpc, 
distance = 9.3E24
area = 4*np.pi*np.square(distance)

def superorbital(x, P, phi):
	return 0.06*np.square(np.cos((np.pi*x/P) + phi)) + 0.01

# def superorbital(x, P, phi, C):
# 	return A*np.square(np.cos((np.pi*x/P) + phi)) + C

def powerlaw(beta, x):
	A, n = beta
	return A*np.power(x, n)

def powerlaw_inv(beta, y):
    A, n = beta
    return np.power(y/A, 1/n)
```

# Hardness Ratio

```python
# HR_mask = (data['XRT Count Rate (0.3-2 keV)'] != 0) * (data['TSTART (MJD)'] > 58400) 
HR_mask = (data['XRT Count Rate (0.3-2.5 keV)'] != 0) * ((data['XRT Count Rate (0.3-10 keV)'] - data['XRT Count Rate (0.3-2.5 keV)'])!= 0)
print(sum(HR_mask))
HR = 1.0-(2*data[HR_mask]['XRT Count Rate (0.3-2.5 keV)']/data[HR_mask]['XRT Count Rate (0.3-10 keV)'])
HR_err = (2/data[HR_mask]['XRT Count Rate (0.3-10 keV)'])*np.sqrt(np.square(data[HR_mask]['XRT Count Rate (0.3-2.5 keV)']*data[HR_mask]['0.3-10 rate +/-']) + np.square(data[HR_mask]['0.3-2.5 rate +/-']))

```

```python
fig = plt.figure(figsize=(12,15))
plt.errorbar(HR, data[HR_mask]['XRT Count Rate (0.3-10 keV)'], yerr=data[HR_mask]['0.3-10 rate +/-'], \
             xerr=HR_err, fmt='none', lw=2)
# fig.get_yaxis().labelpad = 20
plt.xlabel(r'Hardness Ratio (0.3-2.5/2.5-10 keV)', fontsize=32, labelpad=20)
plt.ylabel(r'XRT 0.3-10 keV count rate $(s^{-1})$', fontsize=32)
plt.yscale('log')
# plt.grid()
plt.tight_layout()
# plt.savefig('/Users/sean/Desktop/CircinusULX5/HR_0.3_2.5.pdf')
plt.savefig('/Users/sean/Desktop/CircinusULX5/HR_log_0.3_2.5.pdf')
```

```python
cmap = cm.cool
norm = mpl.colors.Normalize(vmin=np.min(HR), vmax=np.max(HR))
m = cm.ScalarMappable(norm=norm, cmap=cmap)
# print(m.to_rgba(HR))

plt.figure(figsize=(12,8))
plt.scatter((data[HR_mask]['TSTART (MJD)'] + data[HR_mask]['TSTOP (MJD)'])/2, \
             data[HR_mask]['XRT Count Rate (0.3-10 keV)'], c=m.to_rgba(HR), cmap=cmap, s=75)
c = plt.colorbar(m)
plt.errorbar((data[HR_mask]['TSTART (MJD)'] + data[HR_mask]['TSTOP (MJD)'])/2, \
             data[HR_mask]['XRT Count Rate (0.3-10 keV)'], yerr = data[HR_mask]['0.3-10 rate +/-'], fmt='none', ecolor=m.to_rgba(HR), color='black')

plt.xlim((57500,58750))
plt.ylim((0,0.11))
plt.xlabel('MJD')
plt.ylabel(r'XRT 0.3-10 keV count rate $(\mathrm{s^{-1}})$')
plt.tight_layout()
# plt.savefig('/Users/sean/Desktop/XRT_countrate.pdf')

```

```python
cmap = cm.cool
norm = mpl.colors.Normalize(vmin=np.min(HR), vmax=np.max(HR))
m = cm.ScalarMappable(norm=norm, cmap=cmap)
# print(m.to_rgba(HR))

plt.figure(figsize=(12,8))
plt.scatter((data[HR_mask]['TSTART (MJD)'] + data[HR_mask]['TSTOP (MJD)'])/2, \
             data[HR_mask]['XRT Count Rate (0.3-10 keV)'], c=m.to_rgba(HR), cmap=cmap, s=75)
c = plt.colorbar(m)
c.ax.get_yaxis().labelpad = 20
# c.ax.set_ylabel(r'$(H-S)/(H+S)$', rotation=270)
c.ax.set_ylabel('Hardness Ratio', rotation=270)
plt.errorbar((data[HR_mask]['TSTART (MJD)'] + data[HR_mask]['TSTOP (MJD)'])/2, \
             data[HR_mask]['XRT Count Rate (0.3-10 keV)'], yerr = data[HR_mask]['0.3-10 rate +/-'], fmt='none', ecolor=m.to_rgba(HR), color='black')

plt.xlim((58475,58750))
plt.ylim((0,0.09))
plt.xlabel('MJD')
plt.ylabel(r'XRT 0.3-10 keV count rate $(\mathrm{s^{-1}})$')
plt.tight_layout()
plt.savefig('/Users/sean/Desktop/CircinusULX5/2019_countrate_HR.pdf')


```

# Flux comparisons

```python
temp_data = ascii.read('/Users/sean/Desktop/CircinusULX5/CircinusULX5_xrt/Fluxes-Table 1.csv')
flux_mask = temp_data['Flux (erg/cm^2/s) (3-10 keV)'] !=999
fluxes = temp_data[flux_mask]
fluxes
# instruments = ['SWIFT', 'XMM', 'SUZAKU', 'Chandra', 'NuSTAR']
instruments = ['SWIFT', 'XMM', 'NuSTAR']
```

```python
area = 1.076838046437826352360838E51
fig, ax1 = plt.subplots(figsize=(16,8))
i=0
for instrument in instruments:
    temp_mask = fluxes['instrument'] == instrument
    masked_data = fluxes[temp_mask]
    upper_lims = (masked_data['Flux -']/masked_data['Flux (erg/cm^2/s) (3-10 keV)']) > 0.95
    masked_data[upper_lims]['Flux -'] = 0.0
    masked_data[upper_lims]['Flux +'] = 0.0
    ax1.errorbar((masked_data[~upper_lims]['TSTART (MJD)'] + masked_data[~upper_lims]['TSTOP (MJD)'])/2, \
             masked_data[~upper_lims]['Flux (erg/cm^2/s) (3-10 keV)'], \
             yerr = [masked_data[~upper_lims]['Flux -'], masked_data[~upper_lims]['Flux +']],
             ms=20, lw=2,label=instrument, fmt='.', color='C' + str(i))
    ax1.errorbar((masked_data[upper_lims]['TSTART (MJD)'] + masked_data[upper_lims]['TSTOP (MJD)'])/2, \
             masked_data[upper_lims]['Flux (erg/cm^2/s) (3-10 keV)'], \
             yerr = 0, fmt='none', color='C' + str(i), uplims=[True for j in range(np.sum(upper_lims))])
    i+=1
# plt.xscale('log')
ax1.set_yscale('log')
plt.xlim((57530,58810))
# plt.xlim((58470,58750))
ax1.set_xlabel('MJD', fontsize=32)
ax1.set_ylabel(r'Flux 3-10 keV $(\mathrm{erg\ cm^{-2}\ s^{-1}})$',fontsize=32)
ax1.legend()
ax2 = ax1.twinx()
ax2.set_yscale('log')
ax2.set_ylim(ax1.get_ylim()[0]*area, ax1.get_ylim()[1]*area)
ax2.get_yaxis().labelpad = 40
ax2.set_ylabel(r'Luminosity $(\mathrm{erg\ s^{-1}})$', rotation=270, fontsize=32)
for y in ax1.get_yticklabels():
    y.set_fontsize(32)
for y in ax2.get_yticklabels():
    y.set_fontsize(32)
for x in ax1.get_xticklabels():
    x.set_fontsize(32)
ax1.grid()
plt.tight_layout()
# plt.show()
plt.savefig('/Users/sean/Desktop/CircinusULX5/flux_luminosity.pdf')
```
