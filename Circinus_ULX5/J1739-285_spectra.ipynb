```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from astropy.io import ascii
from scipy.optimize import curve_fit

sns.set_context('talk')
sns.set_style("ticks")
sns.set_palette("colorblind")

def line(x,m,b):
    return (m*x) + b
    

def read_QDP(file):
#     Takes in a file and returns an array with the QDP data
    data = [[]]
    i = 0
    for line in file:
        if i >2:
            temp = line.split()
            if temp[0] != 'NO':
                while temp[-1]=='NO':
                    temp = temp[:-1]
                data[-1].append(temp)
            else:
                data.append([])
        i += 1
    for i in range(len(data)):
        for x in data[i]:
            if 'NO' in x:
                print(file)
        data[i] = np.array(data[i]).astype(float).T
    return data

file_root = '/Volumes/LaCie/AstroData/J1739-285/nustar/90501343002_products/'
file_constant = open(file_root + 'constant.qdp', 'r')
file_diskbb = open(file_root + 'tbabs_diskbb.qdp', 'r')
file_diskbb_pl = open(file_root + 'tbabs_diskbb_pl.qdp', 'r')
file_diskbb_bb_pl = open(file_root + 'tbabs_diskbb_bb_pl.qdp', 'r')
file_diskbb_bb_pl_gauss = open(file_root + 'tbabs_diskbb_bb_pl_gauss.qdp', 'r')
file_diskbb_bb_pl_gauss_gauss = open(file_root + 'tbabs_diskbb_bb_pl_gauss_gauss.qdp', 'r')
file_diskbb_bb_pl_gauss_relline = open(file_root + 'tbabs_diskbb_bb_pl_gauss_relline.qdp', 'r')

data_constant = read_QDP(file_constant)
data_diskbb = read_QDP(file_diskbb)
data_diskbb_pl = read_QDP(file_diskbb_pl)
data_diskbb_bb_pl = read_QDP(file_diskbb_bb_pl)
data_diskbb_bb_pl_gauss = read_QDP(file_diskbb_bb_pl_gauss)
data_diskbb_bb_pl_gauss_gauss = read_QDP(file_diskbb_bb_pl_gauss_gauss)
data_diskbb_bb_pl_gauss_relline = read_QDP(file_diskbb_bb_pl_gauss_relline)
```

```python
nsets=2
colors = ['black', 'red']
ls=''
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5,1, sharex=True,  gridspec_kw = {'height_ratios':[3, 1, 1, 1, 1], 'hspace':0}, figsize=(15, 10))
for i in range(nsets):
    ax1.errorbar(data_constant[i][0], data_constant[i][2], xerr=data_constant[i][1], yerr=data_constant[i][3], \
                 ls =ls, color=colors[i], lw=1.5)
    
    ax2.errorbar(data_diskbb_pl[i+nsets][0], data_diskbb_pl[i+nsets][2], xerr=data_diskbb_pl[i+nsets][1], \
                 yerr=data_diskbb_pl[i+nsets][3], ls=ls, color=colors[i], lw=1.0)
    
    ax3.errorbar(data_diskbb_bb_pl[i+nsets][0], data_diskbb_bb_pl[i+nsets][2], xerr=data_diskbb_bb_pl[i+nsets][1], \
                 yerr=data_diskbb_bb_pl[i+nsets][3], ls=ls, color=colors[i], lw=1.0)
    
    ax4.errorbar(data_diskbb_bb_pl_gauss_gauss[i+nsets][0], data_diskbb_bb_pl_gauss_gauss[i+nsets][2], xerr=data_diskbb_bb_pl_gauss_gauss[i+nsets][1], \
                 yerr=data_diskbb_pl[i+nsets][3], ls=ls, color=colors[i], lw=1.0)
    
    ax5.errorbar(data_diskbb_bb_pl_gauss_relline[i+nsets][0], data_diskbb_bb_pl_gauss_relline[i+nsets][2], xerr=data_diskbb_bb_pl_gauss_relline[i+nsets][1], \
                 yerr=data_diskbb_pl[i+nsets][3], ls=ls, color=colors[i], lw=1.0)

ax2.axhline(y=1, lw =2.0, color='cyan')
ax3.axhline(y=1, lw =2.0, color='cyan')
ax4.axhline(y=1, lw =2.0, color='cyan')
ax5.axhline(y=1, lw =2.0, color='cyan')

ax3.set_ylim((0.94,1.06))
ax4.set_ylim((0.94,1.06))
ax5.set_ylim((0.94,1.06))

ax2.set_ylabel('data/model')
ax5.set_xlabel('Energy (keV)')
ax1.set_ylabel(r'$\mathrm{EF_E}$')
ax1.loglog()
plt.tight_layout()
plt.savefig('/Users/sean/Desktop/J1739-285/spectral_fits.pdf')

```
